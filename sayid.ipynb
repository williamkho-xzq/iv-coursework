```python
import pandas as pd
import altair as alt
from IPython.display import display
```

```python
file_path = 'data/movie_dataset.csv'
movies = pd.read_csv(file_path)
movies
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies['genres'] = movies['genres'].apply(lambda x: x.split(',') if isinstance(x, str) else x)  # Adjust this line based on your actual data format
movies['year'] = movies['release_date'].dt.year

movies
```

```python
exploded_genres = movies.explode('genres')
exploded_genres
```

```python
movies_per_year_genre = exploded_genres.groupby(['year', 'genres']).size().reset_index(name='counts')
movies_per_year_genre
```

```python
bar_chart = alt.Chart(movies_per_year_genre).mark_bar().encode(
    x='year:O',  # O for ordinal
    y=alt.Y('counts:Q', stack='zero'),  # Q for quantitative, stack='zero' for stacked bar
    color='genres:N'  # N for nominal
).properties(
    width=800,
    height=400,
    title='Number of Movies Released per Year by Genre'
)

bar_chart.display()
```

```python
movies['first_genre'] = movies['genres'].apply(lambda x: x[0] if isinstance(x, list) else (x[0] if x else None))
movies
```

```python
aggregated_df = movies.groupby(['year', 'first_genre']).agg(
    average_revenue=('revenue', 'mean'),
    average_rating=('vote_average', 'mean')
).reset_index()
aggregated_df
```

```python
# Step 4: Create the bubble plot with Altair
bubble_plot = alt.Chart(aggregated_df).mark_circle().encode(
    x=alt.X('year:O', axis=alt.Axis(title='Year', labelAngle=-45)),
    y=alt.Y('first_genre:N', axis=alt.Axis(title='Genre')),
    size=alt.Size('average_revenue:Q', title='Average Revenue(USD)', scale=alt.Scale(range=[50, 2000])),
    color=alt.Color('average_rating:Q', scale=alt.Scale(scheme='blueorange'), title='Average Rating'),
    tooltip=['year', 'first_genre', 'average_revenue', 'average_rating']
).properties(
    width=600,
    height=400,
    title='Movies: Average Revenue and Rating by Year and First Genre'
)

bubble_plot.display()
```
